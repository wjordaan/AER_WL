### Definitions

```python
def read_ddo(DDOfile):
    '''
    Takes argument DDOfile
    Returns: df_ddo dataframe
    '''
    df_ddo = pd.read_excel(DDOfile, skiprows=10, usecols=94)
    df_ddo["DateTime"] = df_ddo['DATE'] + " " +df_ddo['TIME']           #Create DateTime column
    df_ddo["DateTime"] = pd.to_datetime(df_ddo["DateTime"])                   #Convert to DateTime
    df_ddo = df_ddo.set_index(pd.DatetimeIndex(df_ddo["DateTime"])) #Using DatetimeIndex
    
    df_ddo = df_ddo.drop('DateTime', 1)
    return df_ddo
```

### Import Data

```python
DDOfile = r'9-9~VESTA JOFFRE 102-8-2-40-27~Twilight 2~DDO.xlsx' 

# DDO | BHGE DDO XLSX File
Rig = DDOfile.split("~")[2]
Well = DDOfile.split("~")[1]
Pad = DDOfile.split("~")[0]

Rig = Rig.replace(" ", "_")
Well = Well.replace(" ", "_")
Well = Well.replace("-", "_")
```

### Script

```python
df_ddo = read_ddo(DDOfile)
```

### SQLite

```python
import sqlite3
import pandas as pd
```

Write dataframe to Well_Data database

```python
con = sqlite3.connect('Well_Data.db')
df_ddo.to_sql('df_ddo_'+Well, con=con, if_exists='replace')
```

```python
con = sqlite3.connect('Well_Data.db')
df_test = pd.read_sql_query('SELECT * from df_ddo_'+Well,con)       #Query dataframe
df_test = df_ddo.set_index(pd.DatetimeIndex(df_test["DateTime"]))   #Re-add DateTime index
```
